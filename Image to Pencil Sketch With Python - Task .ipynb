## Beginner Level Task  - Image to Pencil Sketch With Python

### Import the library

```python
import cv2
```

### Read Image

```python
image = cv2.imread("MS.jpg")
cv2.imshow("Dhoni", image)
cv2.waitKey(0)
```

### To read image in rgb format

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

### Converting original image to grayscale

```python
grayscale_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale_img)
```

### Inverting the Image

```python
inverted = 255 - grayscale_img
plt.imshow(inverted)
```

### Blurring the image

Using Gaussian Function of opencv

```python
blurred = cv2.GaussianBlur(inverted, (21,21), 0)
plt.imshow(blurred)
```

### Invert the blurred image

```python
invert_blurr = 255 - blurred
pencil_sketch = cv2.divide(grayscale_img, invert_blurr, scale = 256.0)
plt.imshow(pencil_sketch)
```

```python
rgb_sketch=cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
```
